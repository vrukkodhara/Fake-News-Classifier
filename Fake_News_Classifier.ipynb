```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('train.csv', encoding_errors='ignore', on_bad_lines='skip',engine='python')
df
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0,inplace=True)
df
```

```python
X = df.iloc[:17360,1].values
y = df.iloc[:17360,-1].values
```

```python
from gensim.utils import simple_preprocess
```

```python
corpus=[]
for i in range (X.shape[0]):
  news = simple_preprocess(str(X[i]))
  news = ' '.join(news)
  corpus.append(news)
```

```python
corpus
```

```python
vocabulary_size = 5000
from keras.preprocessing.text import one_hot
one_hotted = [one_hot(word,5000)for word in X ]
```

```python
one_hotted
```

```python
sentence_len = 20
from keras.preprocessing.sequence import pad_sequences
padded_one_hot = pad_sequences(one_hotted,maxlen=sentence_len,padding='post')
padded_one_hot
```

Word Embedding

```python
no_of_features = 40
from tensorflow.keras.layers import Dense,LSTM,Bidirectional,Embedding,Dropout
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Embedding(vocabulary_size,no_of_features,input_length=sentence_len))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units=100)))
model.add(Dropout(0.2))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
X = np.array(padded_one_hot)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train = tf.convert_to_tensor(X_train, dtype=tf.float32)
y_train = tf.convert_to_tensor(y_train, dtype=tf.float32)
X_test = tf.convert_to_tensor(X_test, dtype=tf.float32)
y_test = tf.convert_to_tensor(y_test, dtype=tf.float32)
```

```python
#Bidirectional LSTM
embedding_vector_features=40
model1=Sequential()
model1.add(Embedding(vocabulary_size,embedding_vector_features,input_length=sentence_len))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.3))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model1.summary())
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=2)
```

```python
model1.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=2)
```
